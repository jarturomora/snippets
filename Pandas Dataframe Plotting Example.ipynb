```python
# Initial imports
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Simple line plot
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2018', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
# Simple line plot with 4 series
df = pd.DataFrame(np.random.randn(1000, 5), index=ts.index, columns=["BTC", "ETH", "XRP", "BCH", "LTC"])
df = df.cumsum()
df.plot();
```
